```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from collections import Counter
```

```python
DATA_DIR = r"D:\cell_images_small"
infectedPath = os.path.join(DATA_DIR, "Infected")
uninfectedPath = os.path.join(DATA_DIR, "Uninfected")

infectedImages = os.listdir(infectedPath)
uninfectedImages = os.listdir(uninfectedPath)

print("Infected: ", len(infectedImages))
print("Uninfected: ", len(uninfectedImages))
```

```python
def showImg(classPath, title):
    images = random.sample(os.listdir(classPath), 5)

    plt.figure(figsize=(8,3))
    for i, img in enumerate(images):
        imgPath = os.path.join(classPath, img)
        img = Image.open(imgPath)

        plt.subplot(1,5,i+1)
        plt.imshow(img)
        plt.axis("off")
    plt.suptitle(title)
    plt.show()

showImg(infectedPath, "Infected Samples")
showImg(uninfectedPath, "Uninfected Samples")
```

### Insights
The image need to be resized as they are not of the same size.

```python
infected = []
for img_name in infectedImages[:200]:
    img = Image.open(os.path.join(infectedPath, img_name))
    infected.append(img.size)

print("Unique image size for infected:", set(infected))

uninfected = []
for img_name in uninfectedImages[:200]:
    img = Image.open(os.path.join(uninfectedPath, img_name))
    uninfected.append(img.size)

print("Unique image size for uninfected: ", set(uninfected))
```
